## Análisis de Datos

### 0. Importación de datos



```python
# Paso 0: Importación de datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
# Ahora nos familiarizamos los datos
tienda1.info()
tienda2.info()
tienda3.info()
tienda4.info()
```

```python
tienda1.describe()
```

### 1. Análisis de facturación



En esta sección se calcula el **ingreso total de cada tienda.**

```python
# Paso 1: Análisis de facturación
ingresos_totales_1 = int(tienda1['Precio'].sum())
ingresos_totales_2 = int(tienda2['Precio'].sum())
ingresos_totales_3 = int(tienda3['Precio'].sum())
ingresos_totales_4 = int(tienda4['Precio'].sum())

lista_ingresos = (ingresos_totales_1, ingresos_totales_2, ingresos_totales_3, ingresos_totales_4)
max_ingresos = max(lista_ingresos)
min_ingresos = min(lista_ingresos)

print(f"Los ingresos totales de la Tienda 1 son {ingresos_totales_1}.")
print(f"Los ingresos totales de la Tienda 2 son {ingresos_totales_2}.")
print(f"Los ingresos totales de la Tienda 3 son {ingresos_totales_3}.")
print(f"Los ingresos totales de la Tienda 4 son {ingresos_totales_4}.")
print(f'\nEl máximo de ingresos es {max_ingresos} y corresponde a la tienda {lista_ingresos.index(max_ingresos)+1}.')
print(f'\nEl mínimo de ingresos es {min_ingresos} y corresponde a la tienda {lista_ingresos.index(min_ingresos)+1}.')
```

### 2. Ventas por categoría

Aquí se calcula la cantidad de productos vendidos por categoría en cada tienda.

Primero, se explora el dataframe y se ordena la información.

```python
# Paso 2: Ventas por categoría
# Crear filtros y dfs para cada categoría por cada tienda
def calcular_ventas_por_categoria(df, nombre_tienda):
    categorias = df['Categoría del Producto'].unique()
    ventas = df['Categoría del Producto'].value_counts().to_dict()
    print(f"\nLos valores únicos dentro de la columna 'Categoría del Producto' de {nombre_tienda} son:\n {np.sort(categorias)}\n")
    return ventas

ventas_tienda_1 = calcular_ventas_por_categoria(tienda1, "Tienda 1")
ventas_tienda_2 = calcular_ventas_por_categoria(tienda2, "Tienda 2")
ventas_tienda_3 = calcular_ventas_por_categoria(tienda3, "Tienda 3")
ventas_tienda_4 = calcular_ventas_por_categoria(tienda4, "Tienda 4")

record_ventas_tiendas = pd.DataFrame([ventas_tienda_1, ventas_tienda_2, ventas_tienda_3, ventas_tienda_4],
                                     index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']).fillna(0).astype(int)

# DataFrame Ventas por categoría
print("\nVentas por categoría:\n")
print(record_ventas_tiendas)


# Categorías con mayor número de ventas en cada tienda
print("\nCategorías más vendidas por tienda:\n")
for tienda, row in record_ventas_tiendas.iterrows():
    max_ventas = row.max()
    categorias_max = row[row == max_ventas].index.tolist()
    print(f"- {tienda}: {', '.join(categorias_max)} ({max_ventas} ventas)")

# Categorías con menor número de ventas en cada tienda
print("\nCategorías menos vendidas por tienda:\n")
for tienda, row in record_ventas_tiendas.iterrows():
    min_ventas = row.min()
    categorias_min = row[row == min_ventas].index.tolist()
    print(f"- {tienda}: {', '.join(categorias_min)} ({min_ventas} ventas)")
```

### 3. Calificación promedio de la tienda


```python
# Paso 3: Calificación promedio de la tienda
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}

# Promedios
promedios = pd.Series({
    nombre: pd.to_numeric(df["Calificación"], errors="coerce").mean()
    for nombre, df in tiendas.items()
}).round(2)

# Resultados por tienda
print("Calificación promedio por tienda:")
for nombre, valor in promedios.items():
    print(f"- {nombre}: {valor}")

# Máximo y mínimo con sus tiendas
tienda_max = promedios.idxmax()
tienda_min = promedios.idxmin()

print(f"\nLa calificación máxima fue {promedios.max()} y corresponde a {tienda_max}.")
print(f"La calificación mínima fue {promedios.min()} y corresponde a {tienda_min}.")

resumen_calificaciones = promedios.to_frame("Calificación Promedio")
print("\nResumen:\n")
print(resumen_calificaciones)
```

### 4. Productos más y menos vendidos

```python
# Paso 4: Productos más y menos vendidos
# Paso 4: Productos más vendidos


tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}

print("Productos más vendidos por tienda:\n")
for nombre, df in tiendas.items():
    moda = df["Producto"].mode()
    if len(moda) == 1:
        print(f"- {nombre}: el producto más vendido es '{moda.iloc[0]}'.")
    else:
        productos = ", ".join(moda)
        print(f"- {nombre}: hay empate, los productos más vendidos son: {productos}.")
```

```python
# Paso 4: Productos menos vendidos
print("\nProductos menos vendidos por tienda:\n")
for nombre, df in tiendas.items():
    conteo = df["Producto"].value_counts()
    min_ventas = conteo.min()
    menos_vendidos = conteo[conteo == min_ventas].index.tolist()

    if len(menos_vendidos) == 1:
        print(f"- {nombre}: el producto menos vendido es '{menos_vendidos[0]}'.")
    else:
        productos = ", ".join(menos_vendidos)
        print(f"- {nombre}: hay empate, los productos menos vendidos son: {productos}.")
```

### 5. Envío promedio por tienda

```python
# Paso 5: Envío promedio por tienda

envios_promedio = {}

print("Costo de envío promedio por tienda:\n")
for nombre, df in tiendas.items():
    promedio = df['Costo de envío'].mean()
    envios_promedio[nombre] = promedio
    print(f"- {nombre}: {promedio:.2f}")

tienda_max = max(envios_promedio, key=envios_promedio.get)
tienda_min = min(envios_promedio, key=envios_promedio.get)

print(f"\nEl costo máximo de envío fue {envios_promedio[tienda_max]:.2f} y corresponde a {tienda_max}.")
print(f"El costo mínimo de envío fue {envios_promedio[tienda_min]:.2f} y corresponde a {tienda_min}.")
```

## Gráficos

```python
# Parte 2: Gráficos
import warnings
warnings.filterwarnings("ignore")
import os
output_dir = "graficos_alura_store"
os.makedirs(output_dir, exist_ok=True)

# Paso 1: Gráfico de Ingresos Totales
ingresos = pd.Series(
    [ingresos_totales_1, ingresos_totales_2, ingresos_totales_3, ingresos_totales_4],
    index=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
)

plt.figure(figsize=(6,4))
sns.barplot(x=ingresos.index, y=ingresos.values, palette="Blues_d")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos")
plt.xlabel("Tiendas")
plt.savefig(os.path.join(output_dir, "ingresos_totales.png"))
plt.show()


# Paso 2: Gráficos de Ventas por Categoría (uno por tienda)
for nombre, ventas in zip(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
                          [ventas_tienda_1, ventas_tienda_2, ventas_tienda_3, ventas_tienda_4]):
    plt.figure(figsize=(6,4))
    sns.barplot(x=list(ventas.keys()), y=list(ventas.values()), palette="viridis")
    plt.title(f"Ventas por Categoría - {nombre}")
    plt.ylabel("Cantidad de Ventas")
    plt.xlabel("Categoría del Producto")
    plt.xticks(rotation=45)
    plt.savefig(os.path.join(output_dir, f"ventas_por_categoria_{nombre}.png"))
    plt.show()

# Paso 2: Gráfico de Ventas por Categoría (consolidado/todas las tiendas)
plt.figure(figsize=(10,6))
record_ventas_tiendas.plot(kind="bar", figsize=(10,6))

plt.title("Ventas por Categoría en todas las Tiendas")
plt.ylabel("Cantidad de Ventas")
plt.xlabel("Tiendas")
plt.xticks(rotation=0)
plt.legend(title="Categorías", bbox_to_anchor=(1,1))
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "ventas_por_categoria_consolidado.png"))
plt.show()



# Paso 3: Gráfico de Calificación Promedio por Tienda
plt.figure(figsize=(6,4))
sns.barplot(x=promedios.index, y=promedios.values, palette="magma")
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación Promedio")
plt.xlabel("Tiendas")
plt.ylim(0, 5)
plt.savefig(os.path.join(output_dir, "calificacion_promedio.png"))
plt.show()


# Paso 5: Gráfico de Costo de Envío Promedio por Tienda
envios = pd.Series(envios_promedio)

plt.figure(figsize=(6,4))
sns.barplot(x=envios.index, y=envios.values, palette="magma")
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.xlabel("Tiendas")
plt.savefig(os.path.join(output_dir, "costo_envio_promedio.png"))
plt.show()
```
